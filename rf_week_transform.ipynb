```python
from google.colab import files
import pandas as pd

# Step 1: Upload file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Step 2: Read the CSV
df = pd.read_csv(file_name)

# Step 3: Melt from wide → long format
df_long = df.melt(id_vars=['year', 'week'],
                  var_name='district',
                  value_name='value')

# Step 4: Pivot so weeks are columns (1–53), each district’s weeks horizontal
df_pivot = df_long.pivot_table(
    index=['year', 'district'],
    columns='week',
    values='value',
    aggfunc='first'
)

# Step 5: Flatten the columns and sort
df_pivot = df_pivot.reset_index()
df_pivot = df_pivot.sort_values(['district', 'year'])

# Step 6: Final format — weeks 1–53 columns per district
# This already has one block of 53 columns per district

# Optional: Save
df_pivot.to_csv("districts_weeks_horizontal.csv", index=False)
files.download("districts_weeks_horizontal.csv")
```
